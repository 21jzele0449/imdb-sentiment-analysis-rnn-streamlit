```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences

sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'i am good boy',
        'i am a good developer',
        'understanding the meaniing of words',
        'your videos are good',]
```

```python
sent
```

```python
## define vocabulary size
voc_size = 10000
```

```python
## one hot representation 
one_hot_repr = [one_hot(word, voc_size) for word in sent]
one_hot_repr
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedd_doc = pad_sequences(one_hot_repr , maxlen  = sent_length ,padding = 'pre')
embedd_doc
```

```python
## feaqture representation
dim = 10
```

```python
model=Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_length)) 
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedd_doc)
```
